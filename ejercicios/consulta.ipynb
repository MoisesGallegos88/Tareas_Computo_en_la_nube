```python
s = "hola mundo"

  

# 1. capitalize: Capitaliza la primera letra.

print("capitalize:", s.capitalize())

  

# 2. casefold: Versión en minúsculas ideal para comparaciones insensibles a mayúsculas.

print("casefold:", s.casefold())

  

# 3. center: Centra la cadena en un ancho especificado, rellenando con un caracter.

print("center:", s.center(20, "*"))

  

# 4. count: Cuenta cuántas veces aparece una subcadena.

print("count ('o'):", s.count("o"))

  

# 5. encode: Convierte la cadena a bytes con la codificación dada.

print("encode (utf-8):", s.encode("utf-8"))

  

# 6. endswith: Verifica si la cadena termina con una subcadena.

print("endswith 'mundo':", s.endswith("mundo"))

  

# 7. expandtabs: Reemplaza los tabuladores por espacios.

s_tabs = "hola\tmundo"

print("expandtabs (tabsize=10):", s_tabs.expandtabs(10))

  

# 8. find: Retorna el índice más bajo donde se encuentra la subcadena, o -1 si no existe.

print("find 'm':", s.find("m"))

  

# 9. format: Formatea la cadena utilizando llaves {}.

print("format:", "El número es: {}".format(5))

  

# 10. format_map: Similar a format() pero toma un diccionario.

print("format_map:", "El número es: {num}".format_map({"num": 5}))

  

# 11. index: Igual que find, pero lanza un error si la subcadena no se encuentra.

print("index 'm':", s.index("m"))

  

# 12. isalnum: True si todos los caracteres son alfanuméricos (se eliminan espacios para esta comprobación).

print("isalnum:", s.replace(" ", "").isalnum())

  

# 13. isalpha: True si todos los caracteres son letras (descontando espacios).

print("isalpha:", s.replace(" ", "").isalpha())

  

# 14. isascii: True si todos los caracteres pertenecen al conjunto ASCII.

print("isascii:", s.isascii())

  

# 15. isdecimal: True si la cadena consiste sólo en dígitos decimales.

print("isdecimal ('123'):", "123".isdecimal())

  

# 16. isdigit: True si todos los caracteres son dígitos.

print("isdigit ('123'):", "123".isdigit())

  

# 17. isidentifier: True si la cadena es un identificador válido en Python.

print("isidentifier ('hola'):", "hola".isidentifier())

  

# 18. islower: True si todas las letras están en minúsculas.

print("islower:", s.islower())

  

# 19. isnumeric: True si todos los caracteres son numéricos.

print("isnumeric ('123'):", "123".isnumeric())

  

# 20. isprintable: True si todos los caracteres son imprimibles.

print("isprintable:", s.isprintable())

  

# 21. isspace: True si la cadena está compuesta únicamente por espacios.

print("isspace (' '):", " ".isspace())

  

# 22. istitle: True si la cadena está en formato título (cada palabra comienza con mayúscula).

s_title = "Hola Mundo"

print("istitle:", s_title.istitle())

  

# 23. isupper: True si todas las letras están en mayúsculas.

print("isupper ('HOLA'):", "HOLA".isupper())

  

# 24. join: Une los elementos de un iterable utilizando la cadena como separador.

print("join:", "-".join(["a", "b", "c"]))

  

# 25. ljust: Justifica la cadena a la izquierda en un ancho determinado.

print("ljust:", s.ljust(20, "*"))

  

# 26. lower: Convierte la cadena a minúsculas.

print("lower:", s.lower())

  

# 27. lstrip: Elimina los espacios u otros caracteres de la izquierda.

print("lstrip:", " hola".lstrip())

  

# 28. maketrans y translate: Crea una tabla de transformación y la aplica a la cadena.

tabla = str.maketrans("hola", "HOLA")

print("translate:", s.translate(tabla))

  

# 29. partition: Divide la cadena en una tupla de tres elementos en función de la subcadena.

print("partition (' '):", s.partition(" "))

  

# 30. replace: Reemplaza una subcadena por otra.

print("replace:", s.replace("mundo", "Python"))

  

# 31. rfind: Retorna el índice de la última ocurrencia de la subcadena o -1 si no la encuentra.

print("rfind ('o'):", s.rfind("o"))

  

# 32. rindex: Igual que rfind pero lanza una excepción si no se encuentra.

print("rindex ('o'):", s.rindex("o"))

  

# 33. rjust: Justifica la cadena a la derecha.

print("rjust:", s.rjust(20, "*"))

  

# 34. rpartition: Divide la cadena en tres partes, comenzando por la derecha.

print("rpartition ('l'):", s.rpartition("l"))

  

# 35. rsplit: Divide la cadena desde la derecha usando un separador.

print("rsplit (' '):", s.rsplit(" "))

  

# 36. rstrip: Elimina espacios u otros caracteres del extremo derecho.

print("rstrip:", "hola ".rstrip())

  

# 37. split: Divide la cadena en una lista usando un separador.

print("split (' '):", s.split(" "))

  

# 38. splitlines: Divide la cadena en líneas.

s_lines = "línea1\nlínea2\r\nlínea3"

print("splitlines:", s_lines.splitlines())

  

# 39. startswith: Verifica si la cadena comienza por cierta subcadena.

print("startswith 'hola':", s.startswith("hola"))

  

# 40. strip: Elimina espacios u otros caracteres de ambos extremos.

print("strip:", " hola ".strip())

  

# 41. swapcase: Invierte mayúsculas y minúsculas.

print("swapcase:", s.swapcase())

  

# 42. title: Convierte la cadena a formato título.

print("title:", s.title())

  

# 43. upper: Convierte la cadena a mayúsculas.

print("upper:", s.upper())

  

# 44. zfill: Rellena la cadena con ceros a la izquierda hasta alcanzar el ancho indicado.

print("zfill:", s.zfill(20))

```
